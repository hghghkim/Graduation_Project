```python
# 숫자 순서대로 sort
import re
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
```

```python
# mat file로 저장된 종자의 데이터를 tensorflow적용을 위해 numpy 배열로 변환하여 저장
k=1
BASE_PATH = '/Users/kimhangi/data/Project/old_seeds_data(npy)'
for infile in sorted(glob.glob(r'/Users/kimhangi/data/Project/old_seeds_data(mat)/**/*.mat'), key=numericalSort):
    print("Current File Being Processed is: " + infile)
    with open(infile,"rb") as mat:
        load_mat = scipy.io.loadmat(mat)
        swir_data = load_mat['selected_data3_1']
        file_name = ("{}.npy".format(k))
        k+=1
        np.save(os.path.join(BASE_PATH,file_name),swir_data)
```
